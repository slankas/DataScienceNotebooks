# Word Embeddings


```python
# Download the word2vec model trained from Google news
# If already download, will load into memory
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

```python
# Vector representation for a particular word
vec_computer = wv['computer']
print (vec_computer)
```

```python
# Some similarity scores
wv.most_similar('beautiful')
```

```python
wv.most_similar('paris')
```

```python
# Demonstrate decreasing similarity
pairs = [
    ('car', 'minivan'),   # a minivan is a kind of car
    ('car', 'bicycle'),   # still a wheeled vehicle
    ('car', 'airplane'),  # ok, no wheels, but still a vehicle
    ('car', 'cereal'),    # ... and so on
    ('car', 'communism'),
]
for w1, w2 in pairs:
    print('%r\t%r\t%.2f' % (w1, w2, wv.similarity(w1, w2)))
```

```python
print(wv.most_similar(positive=['car', 'minivan'], topn=5))
```

```python
print(wv.doesnt_match(['fire', 'water', 'land', 'sea', 'air', 'car']))
```
